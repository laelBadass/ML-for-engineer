```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv("salary.csv")
```

```python
df
```

```python
%matplotlib inline
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.scatter(df.year,df.salary,color='green',marker='+')
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['year']],df.salary)
```

```python
reg.predict([[8]])
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
d = pd.read_csv("salary_to_predict.csv")
```

```python
p = reg.predict(d)
```

```python
d['salary'] = p
```

```python
d.to_csv("predicted_salary.csv")
```
